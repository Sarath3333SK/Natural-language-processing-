```python
import pandas as pd
```

```python
dataset=pd.read_csv('mabel.txt',error_bad_lines=False,header=None)
df=pd.DataFrame(columns=["Date"])
df["Date"]=dataset[0]
timemsg=dataset[1].str.split("-",n=1,expand=True)
df["Time"]=timemsg[0]
personmsg=timemsg[1].str.split(":",n=1,expand=True)
df["Person"]=personmsg[0]
df["Message"]=personmsg[1]
df["Person"].value_counts()
df=df.drop(0)
df=df.drop(1)
df=df.dropna()
```
